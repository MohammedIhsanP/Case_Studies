```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

# 1.Read the dataset to the python environment.

```python
filename=pd.read_csv('cars_data.csv')
cars=pd.DataFrame(filename)

cars
```

# 2.Check for the null values present in the dataset.

```python
cars.isna().sum()
```

```python
cars.info()
```

The dataset contains 20 columns and 10000 entries each. From the above information, it is clear that there is no null values present in the dataset

# 3.Plot a bar graph of male vs female buyers participated in the sales.

```python
cars['Buyer Gender'].value_counts()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(4,8))
ax=sns.countplot(cars['Buyer Gender'])
plt.title('Female vs Male Buyers',fontsize=18)
plt.xlabel('Gender',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.tight_layout()

#display the count of each gender
for p in ax.patches:
        ax.text(p.get_x()+.15,p.get_height()-300,str(p.get_height()),fontsize=18,color='white',fontweight='bold')
```

# 4.Find the top 5 cars based on their sales price.

```python
#sorting in descending order of Sale Price
top_car=cars.sort_values(by=['Sale Price'],ascending=False)
```

```python
#replacing old index with new value
top_car=top_car.reset_index(drop=True)
```

```python
top_car[0:5]
```

These are the details of top five cars based on their Sales Price.

# 5.Find the least 5 cars based on their Resell price.

```python
#sorting in ascending order of Resell Price
least_car=cars.sort_values(by=['Resell Price'],ascending=True)
```

```python
#replacing old index with new values 
least_car=least_car.reset_index(drop=True)
```

```python
least_car[0:5]
```
